```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Better visuals
sns.set(style="whitegrid")

# Load sample dataset
df = sns.load_dataset("tips")
df.head()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    style="time"
)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
plt.figure()
sns.barplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("Bar Plot: Average Total Bill by Day and Gender")
plt.show()
```

```python
plt.figure()
sns.violinplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex",
    split=True
)
plt.title("Violin Plot: Bill Distribution by Day and Gender")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure()
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
plt.figure()
sns.swarmplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.title("Swarm Plot: Total Bill by Day")
plt.show()
```
